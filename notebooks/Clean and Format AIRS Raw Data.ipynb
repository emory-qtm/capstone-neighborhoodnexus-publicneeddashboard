```python
#necessary packages to complete basic data cleaning
import numpy as np
import pandas as pd 
import os
```

```python
AIRSdf = pd.read_excel('AIRS Needs Report.xls', sheet_name = 1)
# note, NN plans to have a meeting with United Way to update the way they receive the call information
```

```python
type(AIRSdf)
```

```python
#Clean the data frame - replace mising and NA values with their respective place filler (string or integer) 
cleanDf = AIRSdf.fillna({"AIRS Problem Needs": 'Missing',"AIRS Problem Needs Count" : 0,"Referred and Unmet Services":"Missing", "Referred and Unmet Services Count":0})
cleanDf

#note - for this code (above) to work, the data must be a DataFrame and not a dictionary.
```

```python
#insert new columns that indicate the month and year for current dataframe
cleanDf.insert(0, 'Month', "March") 

cleanDf.insert(1, 'Year', 2021)

#note, users will have to update the input value each month and year. 
```

```python
#Data set with zip codes
cleanDf
```

```python
#Data set without zip codes 
cleanWZip = cleanDf.drop("Search Option - ZIP Code", axis = 1)
cleanWoutZip = cleanWZip.drop("Search Option - ZIP Code Count", axis = 1)
cleanWoutZip
```

```python
# create csv files 

# csv for March data with zip codes 
cleanDf.to_csv("Mar21 AIRS.csv", index = False, encoding = 'utf8')

# csv for March data without zip codes
cleanWoutZip.to_csv("Mar21 AIRS WoutZIP.csv", index = False, encoding = 'utf8')
```

```python
# csv for all data (so far) with zip codes // combined dataset 
cleanDf.to_csv("2021 AIRS Need Report.csv", mode = 'a', header = False, index = False, encoding = 'utf8')

# csv for all data without zip codes 
cleanWoutZip.to_csv("2021 AIRS Need Report WoutZIP.csv", mode = 'a', header = False, index = False, encoding = 'utf8')
```
